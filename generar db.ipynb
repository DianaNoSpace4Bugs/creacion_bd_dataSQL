```python
# Importación de bibliotecas
import pandas as pd
import sqlite3
```

```python
# Creación de dataframes
profes = pd.read_csv('./data/claustro.csv', sep=';')
DS1 = pd.read_csv('./data/listado_DS_clase_1.csv', sep=';')
DS2 = pd.read_csv('./data/listado_DS_clase_2.csv', sep=';')
FS1 = pd.read_csv('./data/listado_FS_clase_1.csv', sep=';')
FS2 = pd.read_csv('./data/listado_FS_clase_2.csv', sep=';')
```

```python
# Creación de dataframe campus
datos_campus = {'nombre_campus':profes['Campus'].unique()}
campus = pd.DataFrame(datos_campus, index=range(1, len(profes['Campus'].unique()) +1))
campus.index.name = "campus_id"
campus
```

```python
# Creación de dataframe promocion
datos_promocion = {'nombre_promocion': [], 'fecha_comienzo': [], 'vertical': [], 'campus_id': []}
for i in [DS1, FS1, DS2, FS2]:
    if len(i['Promoción'].unique()) == 1:
        datos_promocion['nombre_promocion'].append(*i['Promoción'].unique())
    else:
        print('Error. Revisar los datos de \'Promoción\' a mano.')

    if len(i['Fecha_comienzo'].unique()) == 1:
        datos_promocion['fecha_comienzo'].append(*i['Fecha_comienzo'].unique())
    else:
        print('Error. Revisar los datos de \'Fecha_comienzo\' a mano.')

    if 'Proyecto_EDA' in i.columns:
        datos_promocion['vertical'].append('DS')
    else:
        datos_promocion['vertical'].append('FS')

    if i['Campus'].unique() == 'Madrid':
        datos_promocion['campus_id'].append(1)
    else:
        datos_promocion['campus_id'].append(2)

# Ahora generamos el dataframe:
promocion = pd.DataFrame(datos_promocion, index = range(1, len(datos_promocion) + 1))
promocion.index.name = "promocion_id"
promocion
```

```python
# Creación de dataframe claustro:
datos_claustro = {'nombre': profes['Nombre'].to_list(), 
                  'rol': profes['Rol'].to_list(), 
                  'promocion_id': [1, 1, 2, 0, 4, 0, 1, 4, 2, 1]}
datos_claustro
claustro = pd.DataFrame(datos_claustro, index = range(1, len(profes)+1))
claustro.index.name = "profesor_id"
claustro
```

```python
# Creación de dataframe de estudiantes:
datos_estudiantes = {'nombre': [], 'email': [], 'promocion_id': []}
# DS1:
prom = 1
for i in [DS1, FS1, DS2, FS2]:
    [datos_estudiantes['nombre'].append(i['Nombre'][j]) for j in range(len(i['Nombre']))]
    [datos_estudiantes['email'].append(i['Email'][j]) for j in range(len(i['Email']))]
    [datos_estudiantes['promocion_id'].append(prom) for j in range(len(i['Email']))]
    prom += 1


estudiantes = pd.DataFrame(datos_estudiantes, index=range(1, len(datos_estudiantes['nombre'])+1))
estudiantes.index.name = 'alumno_id'
estudiantes

```

```python
# Creación de dataframe evaluación:
datos_evaluacion = {'proyecto_id': [], 'alumno_id': [], 'resultado': []}
# DS1:
al_id = 1
for i in [DS1, FS1, DS2, FS2]:
    for j in i.values:
        for k in range(1, 6):
            if 'Proyecto_FrontEnd' in i.columns:
                datos_evaluacion['proyecto_id'].append(k+5)
            else:
                datos_evaluacion['proyecto_id'].append(k)
            datos_evaluacion['alumno_id'].append(al_id)
            datos_evaluacion['resultado'].append(j[4+k])
        al_id += 1

evaluacion = pd.DataFrame(datos_evaluacion, index=range(1, len(datos_evaluacion['proyecto_id'])+1))
evaluacion.index.name = 'evaluacion_id'
evaluacion
```

```python
# Creación de dataframe proyectos:
datos_proyectos = {'nombre':[]}
[datos_proyectos['nombre'].append(i) for i in DS1.columns[5:]]
[datos_proyectos['nombre'].append(i) for i in FS1.columns[5:]]

proyectos = pd.DataFrame(datos_proyectos, index=range(1, len(datos_proyectos['nombre'])+1))
proyectos.index.name = 'proyecto_id'
proyectos
```

Vamos a generar la estructura de la base de datos en función del esquema
![Imagen](img/esquema.png)

```python
# Creación de la base de datos:

conexion = sqlite3.connect('escuela.db') 
c = conexion.cursor()

# Creación de tablas
c.execute('''
            CREATE TABLE IF NOT EXISTS campus (
            campus_id integer PRIMARY KEY,
            nombre_campus TEXT
            );
          ''')

c.execute('''
            CREATE TABLE promocion (
            promocion_id INTEGER PRIMARY KEY,
            nombre_promocion TEXT,
            fecha_comienzo DATE,
            vertical TEXT,
            campus_id INT NOT NULL,
            FOREIGN KEY (campus_id) REFERENCES campus(campus_id)
            )
          ''')

c.execute('''
            CREATE TABLE estudiantes (
            alumno_id INTEGER PRIMARY KEY,
            nombre TEXT,
            email TEXT,
            promocion_id INT NOT NULL,
            FOREIGN KEY (promocion_id) REFERENCES promocion(promocion_id)
            )
          ''')

c.execute('''
            CREATE TABLE evaluacion (
            evaluacion_id INTEGER PRIMARY KEY,
            proyecto_id INTEGER,
            alumno_id INTEGER NOT NULL,
            resultado TEXT,
            FOREIGN KEY (proyecto_id) REFERENCES proyectos(proyecto_id),
            FOREIGN KEY (alumno_id) REFERENCES estudiantes(alumno_id)
            )
          ''')

c.execute('''
            CREATE TABLE claustro (
            profesor_id INTEGER PRIMARY KEY,
            nombre TEXT,
            rol TEXT,
            promocion_id INTEGER,
            FOREIGN KEY (promocion_id) REFERENCES promocion(promocion_id)
            )
          ''')

c.execute('''
            CREATE TABLE IF NOT EXISTS proyectos (
            proyecto_id integer PRIMARY KEY,
            nombre TEXT
            );
          ''')

conexion.commit()
```

```python
# Copia de datos de dataframe a SQL
promocion.to_sql('promocion', conexion, if_exists='append')
campus.to_sql('campus', conexion, if_exists='append')
claustro.to_sql('claustro', conexion, if_exists='append')
estudiantes.to_sql('estudiantes', conexion, if_exists='append')
evaluacion.to_sql('evaluacion', conexion, if_exists='append')
proyectos.to_sql('proyectos', conexion, if_exists='append')
```

```python
# Cerrar conexión
conexion.close()
```
